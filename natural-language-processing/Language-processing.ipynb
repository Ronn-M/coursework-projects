```python
# defining dataframe main primary (word list) and secondary columns (Paragraph to word frequency)
DfColumn1 = 'WORD'
DfColumn2 = 'FREQUENCY' 

# Funtions
###################################################################################################################

# function to create dataframe from word tokens and add the Total 
def create_df(word_token):
    
    df = DataFrame(word_token, columns = [DfColumn1]) 
    df = df.append({DfColumn1:'Total'},ignore_index = True)
    return df
###################################################################################################################

# function to add columns from .csv dictionary column list
def add_columns(Tokenized_df,Dict_columnlist):
    
    for col_count in (Dict_columnlist):
        Tokenized_df[col_count] = 0
    return Tokenized_df
###################################################################################################################

# function to add a single column to dataframe
def add_column(Tokenized_df,column):
    Tokenized_df[column] = 0
    return Tokenized_df
###################################################################################################################

# function for dataframe population for comparing the availablity of word token in dictionary
def df_populate(dataframe,dictionary):
    
    # defining primary column for loop
    dfmain_col  = dataframe.columns[0]
    
    # loop for selecting active row to populate
    for cell in range(len(dataframe[dfmain_col])):  
    
        # loop for selecting column to populate
        for col in (dictionary):   
        
            # loop for current row and corresponding column population
            for active_cell in range(len(dictionary[col])): 
            
                # comparison statements for word and dictionary column contents
                if dataframe[dfmain_col][cell] == dictionary[col][active_cell]:
                    dataframe[col][cell]        = 1                  
                else:
                    pass
    return dataframe      
###################################################################################################################

# function for dataframe population by comparing availablity of word token in Textblock/paragraph
def df_popcol(dataframe, textblock):
    
    # loop for selecting active row to populate
    for cell in range(len(WordList_Tokens)):    
            
        
        counter = 0
        word = dataframe[dataframe.columns[0]][cell]
        
        # loop for textblock iteration
        for count in range (len(textblock)):
            
            # comparison statements for word and textblock contents
            if word in textblock[count]: 
                counter += 1    
            else:
                pass
            
        dataframe[dataframe.columns[1]][cell] = counter  
                
    return dataframe  
###################################################################################################################

# Function for column totals
def Totals(dataframe):
    
    # defining primary column for data collection
    dfmain_col  = Word_df.columns[0]
    
    # total rows in dataframe
    Total = len(dataframe[dfmain_col])

    #subtract 1 to account for the 0 index
    Total -= 1
    
    # Loop to select column to be totaled and record total
    for count in (dataframe.columns):  
        
        if count == dfmain_col:
            pass
        else: 
            dataframe[count][Total] = np.sum(dataframe[count])
    return dataframe
```

```python
# defining dataframe main primary columns
DfColumn1 = 'TAG'

# Functions
###################################################################################################################

# function to create dataframe from word tokens 
def create_df(word_token):
    
    df = DataFrame(word_token, columns = [DfColumn1]) 
    df = df.append({DfColumn1:'Total'},ignore_index = True)
    return df
###################################################################################################################

# function to add columns from .csv dictionary column list
def add_columns(Tokenized_df,Dict_columnlist):
    
    for col_count in (Dict_columnlist):
        Tokenized_df[col_count] = 0
    return Tokenized_df
###################################################################################################################

# function for dataframe population by comparing availablity of word token in dictionary
def df_populate(dataframe,taglist,wordtoken):
    
    # defining primary column for loop
    dfmain_col  = DfColumn1
    
    # loop for selecting active row to populate
    for cell in range(len(dataframe[dfmain_col])):  
    
        # loop for selecting column to populate
        for col in (taglist):   
        
            # loop for current row and corresponding column population
            for active_cell in range(len(taglist)): 
            
                # comparison statements for word and dictionary column contents
                if dataframe[dfmain_col][cell] == taglist[active_cell]:
                    dataframe[wordtoken[active_cell]][cell]        = 1                  
                else:
                    pass
    return dataframe   
###################################################################################################################

# function for culculation of frequency occurance - "#1"
def frequency_Totals(dataframe,wordtoken):
    
    #defining primary column for loop
    dfmain_col  = dataframe.columns[0]
    
    #total rows in dataframe
    Total_dfrows = len(dataframe[dfmain_col])

    #subtract 1 to account for the 0 index
    Total_dfrows -= 1

    # Loop to select column to be totaled and record total
    for count in (wordtoken):  
        dataframe[count][Total_dfrows] = np.sum(dataframe[count])
    return dataframe
###################################################################################################################
```
