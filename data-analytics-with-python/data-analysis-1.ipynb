```python
import matplotlib.pyplot as plt
%matplotlib inline


from scipy.stats import norm
from numpy.random import randn

import numpy as np
import pandas as pd
```

```python
Age = np.random.randint(18, high = 30, size = 25)
AgeInt = Age
Time = np.random.randint(1, high = 20, size = 100)



WATCHING_TELEVISION =  Time[0:25]
WT = WATCHING_TELEVISION

USING_SOCIAL_MEDIA = Time[25:50]
USM = USING_SOCIAL_MEDIA

TAKING_AN_ONLINE_CLASS = Time[50:75]
TAOC = TAKING_AN_ONLINE_CLASS

TAKING_A_MODULAR_CLASS = Time[75:100]
TAMC = TAKING_A_MODULAR_CLASS
```

```python
df = pd.DataFrame(Age,
columns = ['Age'])
index = Age

df['WATCHING TELEVISION'] = WATCHING_TELEVISION
df['USING SOCIAL MEDIA'] = USING_SOCIAL_MEDIA
df['TAKING_AN_ONLINE_CLASS'] = TAKING_AN_ONLINE_CLASS
df['TAKING_A_MODULAR_CLASS'] = TAKING_A_MODULAR_CLASS

df.sort_values(['Age'])
```

```python
ax = df.sort_values(['Age']).plot(kind = 'bar', figsize = (15, 10))
```

```python
ax = df.plot(kind = 'line', figsize = (15, 10))
plt.show()
```

```python
plt.scatter(x = Age,y = WATCHING_TELEVISION, label = "WATCHING TELEVISION")
plt.scatter(x = Age,y = USING_SOCIAL_MEDIA, label = "USING SOCIAL MEDIA")
plt.scatter(x = Age,y = TAKING_AN_ONLINE_CLASS, label = "TAKING AN ONLINE CLASS")
plt.scatter(x = Age,y = TAKING_A_MODULAR_CLASS, label = "TAKING A MODULAR CLASS")
plt.legend(bbox_to_anchor = (0., 1.02, 1., .102), loc = 3,
           ncol = 2, mode = "expand", borderaxespad = 0)
plt.show()
```

```python
data = np.concatenate((Age, WATCHING_TELEVISION, USING_SOCIAL_MEDIA,TAKING_AN_ONLINE_CLASS,TAKING_A_MODULAR_CLASS))
plt.boxplot(data)
plt.show()
```
